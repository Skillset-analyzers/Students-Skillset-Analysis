# **Endoing Techniques to follow**

**Binary Encoder** needed in *gender, dropper, firstgrad, volunteer, clubs, software, hardware, ever_pitched*

**Label Encoder** (Ordinal Data) needed in *year, branch, type_sch10, per_sch10,loc_sch10, type_sch12, per_sch12, loc_sch12, no_of_lang, dsa, github, communities, competitions*

**One Hot Encoding/User Defined Function** needed in *domain, add_Skills, achievements*

Reference: https://www.geeksforgeeks.org/feature-encoding-techniques-machine-learning/

# Binary Encoding Function

```python
def bin_Encoder(colname : str):
  print("Unique Values in %s are " % colname)
  print(sorted(df[colname].unique()))
  df[colname] = df[colname].apply(
    lambda x: 1 if x == sorted(df[colname].unique())[0] else (0 if x == sorted(df[colname].unique())[1] else None))
  print("Unique Values in %s are " % colname)
  print(sorted(df[colname].unique()))
```

```python
columns_for_binary=['gender', 'dropper', 'firstgrad', 'volunteer', 'clubs', 'software', 'hardware', 'ever_pitched']
for i in columns_for_binary:
  bin_Encoder(i)
  print('\n')
```

# One Hot Encoding

**One Hot Encoding of "Domain"**

```python
df[['Domain_Development','Domain_Research','Domain_Designing','Domain_Competetive_Coding']] = df['domain'].apply(lambda x: pd.Series(str(x).split(";")))
df['domain_None']=0
i=0
for ele in df['domain']:
  # print( ele.split(";"))
  if 'Development' in ele.split(";"):
    df.at[i,'Domain_Development']=1
  else:
      df.at[i,'Domain_Development']=0
  if 'Research' in ele.split(";"):
    df.at[i,'Domain_Research']=1
  else:
      df.at[i,'Domain_Research']=0  
  if 'Designing' in ele.split(";"):
    df.at[i,'Domain_Designing']=1
  else:
      df.at[i,'Domain_Designing']=0  
  if 'Competetive_Coding' in ele.split(";"):
    df.at[i,'Domain_Competetive_Coding']=1
  else:
      df.at[i,'Domain_Competetive_Coding']=0  
  if 'None' in ele.split(";"):
    df.at[i,'domain_None']=1
  else:
      df.at[i,'domain_None']=0      
  i+=1   
             
df
```

**One Hot Encoding of "add_skills"**

```python
df[['add_skills_Photography','add_skills_Video Editing','add_skills_Sports','add_skills_Cultural Activities','add_skills_None']] = df["add_skills"].apply(lambda x: pd.Series(str(x).split(";")))
i=0
for ele in df['add_skills']:
  # print( ele.split(";"))
  if 'Photography' in ele.split(";"):
    df.at[i,'add_skills_Photography']=1
  else:
      df.at[i,'add_skills_Photography']=0
  if 'Video Editing' in ele.split(";"):
    df.at[i,'add_skills_Video Editing']=1
  else:
      df.at[i,'add_skills_Video Editing']=0  
  if 'Sports' in ele.split(";"):
    df.at[i,'add_skills_Sports']=1
  else:
      df.at[i,'add_skills_Sports']=0  
  if 'Cultural Activities' in ele.split(";"):
    df.at[i,'add_skills_Cultural Activities']=1
  else:
      df.at[i,'add_skills_Cultural Activities']=0  
  if 'None' in ele.split(";"):
    df.at[i,'add_skills_None']=1
  else:
      df.at[i,'add_skills_None']=0      
  i+=1   
df             
```

**One Hot Encoding of "Achievements"**

```python
df[['achievements_Scholarships','achievements_Hackathons','achievements_Coding Competitions','achievements_Research Papers','achievements_Projects','achievements_None']] = df["achievements"].apply(lambda x: pd.Series(str(x).split(";")))
i=0
for ele in df['achievements']:
  # print( ele.split(";"))
  if 'Scholarships' in ele.split(";"):
    df.at[i,'achievements_Scholarships']=1
  else:
      df.at[i,'achievements_Scholarships']=0
  if 'Hackathons' in ele.split(";"):
    df.at[i,'achievements_Hackathons']=1
  else:
      df.at[i,'achievements_Hackathons']=0    
  if 'Coding Competitions' in ele.split(";"):
    df.at[i,'achievements_Coding Competitions']=1
  else:
      df.at[i,'achievements_Coding Competitions']=0  
  if 'Research Papers' in ele.split(";"):
    df.at[i,'achievements_Research Papers']=1
  else:
      df.at[i,'achievements_Research Papers']=0  
  if 'Projects' in ele.split(";"):
    df.at[i,'achievements_Projects']=1
  else:
      df.at[i,'achievements_Projects']=0    
  if 'None' in ele.split(";"):
    df.at[i,'achievements_None']=1
  else:
      df.at[i,'achievements_None']=0    
  i+=1      
df              
```
